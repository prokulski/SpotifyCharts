```python
import geopandas as gpd
import pandas as pd
```

```python
# dane mapowe z https://www.naturalearthdata.com/downloads/110m-cultural-vectors/
world = gpd.read_file('/home/lemur/Work/data/mapy_shp/world/ne_110m_admin_0_countries.shp')

# żeby liczyć odległość w metrach
world.to_crs(epsg=3310,inplace=True)

# środki państw
world['center_point'] = world.centroid
```

```python
distance_df = pd.DataFrame()

for _, row in world.iterrows():
    d = world.center_point.distance(row['center_point']) / 1000 # odległość w km
    d_df = pd.DataFrame({'from': row['ISO_A2'],
                        'to': world.ISO_A2,
                        'distance': d})
    distance_df = distance_df.append(d_df)

distance_df = distance_df[distance_df['distance'] != 0]
distance_df.reset_index(inplace=True, drop=True)
```
